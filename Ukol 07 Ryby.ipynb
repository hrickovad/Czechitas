Ryby

V souboru Fish.csv najdeš informace o rybách z rybího trhu:

- délku (vertikální - Length1, diagonální - Length2 a úhlopříčnou - Length3),
- výšku,
- šířku,
- živočišný druh ryby,
- hmnotnost ryby.

Vytvoř regresní model, který bude predikovat hmnotnost ryby na základě její diagonální délky (sloupec Length2). 

Zkus přidat do modelu výšku ryby (sloupec Height) a porovnej,jak se zvýšila kvalita modelu. 

Nakonec pomocí metody target encoding zapracuj do modelu živočišný druh ryby.

```python
import requests
import pandas as pd
import statsmodels.formula.api as smf

```

```python
r = requests.get("https://raw.githubusercontent.com/pesikj/progr2-python/master/data/Fish.csv")
with open("Fish.csv", "wb") as f:
  f.write(r.content)
```

```python
fish = pd.read_csv("Fish.csv")
```

```python
fish
```

```python
fish['Species'].unique()
```

```python
model_fish= smf.ols(formula = "Weight ~ Length2", data= fish)
```

```python
result = model_fish.fit() 
result.summary()
```

```python
model_fish_height= smf.ols(formula = "Weight ~ Length2 + Height", data= fish)
```

```python
result_height = model_fish_height.fit() 
result_height.summary()

# koeficient determinace sa zvysil na 0.875 - model je presnejsi, odhad vahy ryby bude presnejsi
```

```python
species_width_mean =  fish.groupby("Species")['Width'].mean()
```

```python
fish['Width_mean'] = fish['Species'].map(species_width_mean)
```

```python
fish
```

```python
model_fish_width = smf.ols(formula = "Weight ~ Length2 + Height + Width_mean", data= fish)
```

```python
result_width = model_fish_width.fit() 
result_width.summary()

# model sa spresnil na 0.91 , asi naozaj hrá dosť velkú rolu vo váhe ryby šírka (asi sa myslí tloušťka) ryby
# keď si predstavím štiku a kapra, majú takmer rovnakú dľžku tela v dospelosti, tak štika váži pri svojej veľkosti 1,5 kg,kapr 4 kg
```
